# Why Data Visualization ?

#### Data visualization is a quick, easy way to convey concepts in a universal manner.

# What Is Data Visualization ?

#### Data visualization is the graphical representation of information and data.
#### Visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.

# What Is Matplotlib ?

#### Matplotlib is a plotting libary for the python programming language and its numerical mathematics extention NumPy.
#### Matplotlib is a python library used for Data Visualization.
#### Matplotlib is 2D and 3D plotting python library.

#  Matplotlib Bar Plot

```python
import matplotlib.pyplot as plt
x=["python" , "c" , "cpp" , "Java"]
y=[85,70,60,82]

plt.title("wscube")
c = ["y" , 'g' , "r" , 'b']
plt.bar(x,y , color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]

plt.xlabel("Language", fontsize=15)  
plt.ylabel("No.", fontsize=15)      
plt.title("WSCube", fontsize=15)  
c=["y", 'b' ,"m" ,'g']
plt.bar(x, y, width=0.4 , color=c, edgecolor="r",linewidth=5,linestyle=":",alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]

plt.xlabel("Language", fontsize=15)  
plt.ylabel("No.", fontsize=15)      
plt.title("WSCube", fontsize=15)  
c=["y", 'b' ,"m" ,'g']
plt.bar(x, y, width=0.4 , color=c, edgecolor="r",label="popularity")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20,30,40,50]
plt.xlabel("Language", fontsize=15)  
plt.ylabel("No.", fontsize=15)      
plt.title("WSCube", fontsize=15)  
c=["y", 'b' ,"m" ,'g']
plt.bar(x, y, width=0.4 , color="r",label="popularity")
plt.bar(x, z, width=0.4 , color="y",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]
width = 0.2
p = np.arange(len(x))


p1 = [j + width for j in p]

plt.xlabel("Language", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("WSCube", fontsize=15)

# Corrected bar plotting
plt.bar(p, y, width=width, color="r", label="Popularity")
plt.bar(p1, z, width=width, color="y", label="Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20,30,40,50]

plt.xlabel("Language", fontsize=15)  
plt.ylabel("No.", fontsize=15)      
plt.title("WSCube", fontsize=15) 


plt.barh(x, y , color="m",label="popularity")
plt.barh(x, y , color="b",label="popularity1")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.title("Scatter plot", fontsize=15)
plt.show
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

plt.scatter(day, no)
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
colors=["r","y","g","b","r","g","r"]
plt.scatter(day, no,c=colors)
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
colors=["r","y","g","b","r","g","r"]
size=[400,200,400,300,200,100,600]
plt.scatter(day, no,c=colors,s=size)
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
colors=["r","y","g","b","r","g","r"]
size=[400,200,400,300,200,100,600]
plt.scatter(day, no,c=colors,s=size,alpha=0.4,marker="*")
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
colors = [1, 2, 3, 4, 5, 6, 7]  # Numeric values instead of color names
size = [400, 200, 400, 300, 200, 100, 600]

plt.scatter(day, no, c=colors, s=size, cmap="BrBG")
plt.colorbar()  # Color bar now corresponds to the numeric values
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
colors = [1, 2, 3, 4, 5, 6, 7]  
size = [400, 200, 400, 300, 200, 100, 600]

plt.scatter(day, no, c=colors, s=size, cmap="viridis", alpha=0.7)  # Fixed alpha parameter
t = plt.colorbar()  
t.set_label("ColorBar", fontsize=15)  # Fixed set_label typo
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```

# Stem Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 7, 3, 8]

plt.stem(x, y)
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 7, 3, 8]

plt.stem(x, y,linefmt=":", markerfmt="r+")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 7, 3, 8]

plt.stem(x, y,linefmt=":", markerfmt="ro",bottom=9)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 7, 3, 8]

plt.stem(x, y,linefmt=":", markerfmt="ro",bottom=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 7, 3, 8]

plt.stem(x, y,linefmt=":", markerfmt="ro",bottom=0,basefmt="g")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 7, 3, 8]

# Corrected 'lable' to 'label'
plt.stem(x, y, linefmt=":", markerfmt="ro", bottom=0, basefmt="g", label="python")

plt.legend()
plt.show()
```

# Stack Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
area = [1, 2, 3, 4, 5]

plt.stackplot(x, area)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
area = [1,2, 3, 4, 5]
area1 = [2, 3, 4, 5, 6]
area2 = [1, 3, 2, 4, 2]

plt.stackplot(x, area, area1, area2, labels=["area", "area1", "area2"])

plt.legend(loc="upper left")
plt.show()
```

# Step Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

plt.step(x, y)

plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]


plt.step(x, y, color="r", marker="o")

plt.title("Python")
plt.xlabel("X Axis")  # Corrected xlabel
plt.ylabel("Y Axis")  # Corrected ylabel
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]


plt.step(x, y, color="r", marker="o",mfc="g")

plt.title("Python")
plt.xlabel("X Axis")  
plt.ylabel("Y Axis")  
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]


plt.step(x, y, color="r", marker="o", markerfacecolor="g", label="python")

plt.title("Python")
plt.xlabel("X Axis")  
plt.ylabel("Y Axis") 
plt.legend(loc=2)
plt.grid(True)
plt.show()
```

# Fill_Between Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.fill_between(x, y)

plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1,2,3,4,5]

plt.plot(x,y, color="r")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1,2,3,4,5]

plt.plot(x,y, color="r")

plt.fill_between(x,y)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1,2,3,4,5]

plt.plot(x,y, color="r")

plt.fill_between(x=[2,4],y1=2,y2=4)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1,2,3,4,5]

plt.plot(x,y, color="r")

plt.fill_between(x,y,color="g")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Subplot Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.subplot(1, 1, 1)  

plt.plot(x, y)
plt.show()

```

```python
import matplotlib.pyplot as plt

xx = [1, 2, 3, 4]
y = [1, 2, 3, 4]

plt.plot(xx, y, color="r")


plt.pie([1], colors=["r"])

plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]


plt.subplot(2, 2, 1)
plt.plot(x, y, color="r")


plt.subplot(2, 2, 2)
plt.pie([1], colors=["r"])


x1 = [10, 20, 30, 40]
plt.subplot(2, 2, 3)
plt.pie(x1, colors=["g", "b", "y", "c"])

x2 = ["a", "s", "d", "f"]
y2 = [2, 4, 6, 8]
plt.subplot(2, 2, 4)
plt.bar(x2, y2)


plt.show()
```
