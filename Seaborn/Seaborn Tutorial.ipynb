# What's the difference between countplot and barplot ??

### Countplot plots the count of the number of records by category.
### Barplot plots a value or metric for each category.

# Multiple Plot

## Facet Grid

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset("tips")
var
```

```python
fg = sns.FacetGrid(var, col="sex",hue="day")
fg.map(plt.scatter, "total_bill", "tip")
plt.show()
```

```python
fg = sns.FacetGrid(var, col="sex",hue="day")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var, col="sex",hue="sex")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var, col="sex",hue="sex")
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var, col="sex",hue="sex",palette='summer')
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```
